```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv('C:/Users/yasse/Documents/Report MOnte carlo/archive (2)/data.csv')
df.head()
```

```python


# Select features and target
features = ['sqft_living', 'bedrooms', 'bathrooms']
target = 'price'

X = df[features].values
y = df[target].values

# Normalize features
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Add bias (intercept) term
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape: (n_samples, n_features + 1)

# Initialize parameters
theta = np.random.randn(X_b.shape[1])
learning_rate = 0.01
n_iterations = 1000
m = len(X_b)

# Store loss values for plotting
losses = []

# Gradient Descent Loop
for iteration in range(n_iterations):
    gradients = (2/m) * X_b.T @ (X_b @ theta - y)
    theta -= learning_rate * gradients
    loss = np.mean((X_b @ theta - y) ** 2)
    losses.append(loss)

# Final model parameters
print("Final parameters (theta):", theta)

# Regression equation
print(f"price = {theta[0]:.2f} + {theta[1]:.2f}*sqft_living + {theta[2]:.2f}*bedrooms + {theta[3]:.2f}*bathrooms")

# Plot loss over iterations
plt.plot(losses)
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.title("Loss over Time (Gradient Descent)")
plt.grid()
plt.show()
```
